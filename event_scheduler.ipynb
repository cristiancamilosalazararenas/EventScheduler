```python
class Nodo:
    def __init__(self, clave, color = 'RED'):
        self.color = color
        self.clave = clave
        self.izq = None
        self.der = None
        self.padre = None

    class arbolRN:
        #Constructor de los arboles rojinegros
        def __init__(self):
            self.nil = None(None, 'BLACK')
            self.raiz = self.nil
        

        def insertar(self, clave):
            if self.raiz == self.nill:
                self.raiz = Nodo(clave, 'BLACK')
                self.raiz.izq = self.raiz.der = self.nil
            else:
                self.__insertarRecursivo(self.raiz, clave)
        def __insertarRecursivo(self, nodoActual, clave):
            if nodoActual.clave > clave:
                if nodoActual.izq == self.nil:
                    nodoActual.izq = Nodo(clave)
                    nodoActual.izq.padre = nodoActual
                    nodoActual.izq.izq = nodoActual.der.der = self.nil #Tanto nodoActual.izq.izq como nodoActual.der.der van a valer self.nil
                    self.__arreglarInsertar(nodoActual.izq)
                else:
                    self.__insertarRecursivo(nodoActual.izq, clave)
            else:
                if nodoActual.der == self.nil:
                    nodoActual.der = Nodo(clave)
                    nodoActual.der.padre = nodoActual
                    nodoActual.der.der = nodoActual.der.izq = self.nil
                    self.__arreglarInsertar(nodoActual.der)
                else:
                    self.__insertarRecursivo(nodoActual.der, clave)
        def __arreglarInsertar(self, nodoActual):
            while nodoActual.padre and nodoActual.padre.color == 'RED':
                if nodoActual.padre == nodoActual.padre.padre.izq:
                    uncle = nodoActual.padre.padre.der
                    if uncle.color == 'RED':
                        uncle.color = 'BLACK'
                        nodoActual.padre.color = 'BLACK'
                        nodoActual.padre.padre.color = 'RED'

                        nodoActual = nodoActual.padre.padre
                    else:
                        if nodoActual == nodoActual.padre.der:
                            nodoActual = nodoActual.padre
                            self.__rotarIzq(nodoActual)
                        nodoActual.padre.color = 'BLACK'
                        nodoActual.padre.padre.color = 'RED'
                        self.__rotarDer(nodoActual.padre.padre)
                else:
                    uncle = nodoActual.padre.padre.izq
                    if uncle.color == 'RED':
                        uncle.color = 'BLACK'
                        nodoActual.padre.color = 'BLACK'
                        nodoActual.padre.padre.color = 'RED'

                        nodoActual = nodoActual.padre.padre
                    else:
                        if nodoActual == nodoActual.padre.izq:
                            nodoActual = nodoActual.padre
                            self.__rotarDer(nodoActual)
                        nodoActual.padre.color = 'BLACK'
                        nodoActual.padre.padre.color = 'RED'
                        self.__rotarIzq(nodoActual)

        def __rotarIzq(self, nodoActual):
            nodoCambio = nodoActual.der
            nodoActual.der = nodoCambio.izq

            if nodoCambio.izq != self.nil:
                nodoCambio.izq.padre = nodoActual
            
            nodoCambio.padre = nodoActual.padre

            if nodoActual.padre is None:
                self.raiz = nodoCambio
            elif nodoActual.padre.izq == nodoActual:
                nodoActual.padre.izq = nodoCambio
            else:
                nodoActual.padre.der = nodoCambio

            nodoCambio.izq = nodoActual
            nodoActual.padre = nodoCambio

            #SOLO FALTA ROTAR HACIA LA DERECHA
```
