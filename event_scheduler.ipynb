```python
class Nodo:
    def __init__(self, clave, color='RED'):
        self.color = color
        self.clave = clave
        self.izq = None
        self.der = None
        self.padre = None

class ArbolRN:
    # Constructor del a rbol rojinegro
    def __init__(self):
        self.nil = Nodo(None, 'BLACK')
        self.raiz = self.nil

    def insertar(self, clave):
        if self.raiz == self.nil:
            self.raiz = Nodo(clave, 'BLACK')
            self.raiz.izq = self.raiz.der = self.nil
        else:
            self.__insertarRecursivo(self.raiz, clave)

    def __insertarRecursivo(self, nodoActual, clave):
        if nodoActual.clave > clave:
            if nodoActual.izq == self.nil:
                nodoActual.izq = Nodo(clave)
                nodoActual.izq.padre = nodoActual
                nodoActual.izq.izq = nodoActual.izq.der = self.nil
                self.__arreglarInsertar(nodoActual.izq)
            else:
                self.__insertarRecursivo(nodoActual.izq, clave)
        else:
            if nodoActual.der == self.nil:
                nodoActual.der = Nodo(clave)
                nodoActual.der.padre = nodoActual
                nodoActual.der.izq = nodoActual.der.der = self.nil
                self.__arreglarInsertar(nodoActual.der)
            else:
                self.__insertarRecursivo(nodoActual.der, clave)

    def __arreglarInsertar(self, nodoActual):
        while nodoActual.padre and nodoActual.padre.color == 'RED':
            if nodoActual.padre == nodoActual.padre.padre.izq:
                uncle = nodoActual.padre.padre.der
                if uncle.color == 'RED':
                    uncle.color = 'BLACK'
                    nodoActual.padre.color = 'BLACK'
                    nodoActual.padre.padre.color = 'RED'
                    nodoActual = nodoActual.padre.padre
                else:
                    if nodoActual == nodoActual.padre.der:
                        nodoActual = nodoActual.padre
                        self.__rotarIzq(nodoActual)
                    nodoActual.padre.color = 'BLACK'
                    nodoActual.padre.padre.color = 'RED'
                    self.__rotarDer(nodoActual.padre.padre)
            else:
                uncle = nodoActual.padre.padre.izq
                if uncle.color == 'RED':
                    uncle.color = 'BLACK'
                    nodoActual.padre.color = 'BLACK'
                    nodoActual.padre.padre.color = 'RED'
                    nodoActual = nodoActual.padre.padre
                else:
                    if nodoActual == nodoActual.padre.izq:
                        nodoActual = nodoActual.padre
                        self.__rotarDer(nodoActual)
                    nodoActual.padre.color = 'BLACK'
                    nodoActual.padre.padre.color = 'RED'
                    self.__rotarIzq(nodoActual.padre.padre)

    def __rotarIzq(self, nodoActual):
        nodoCambio = nodoActual.der
        nodoActual.der = nodoCambio.izq
        if nodoCambio.izq != self.nil:
            nodoCambio.izq.padre = nodoActual
        nodoCambio.padre = nodoActual.padre

        if nodoActual.padre is None:
            self.raiz = nodoCambio
        elif nodoActual == nodoActual.padre.izq:
            nodoActual.padre.izq = nodoCambio
        else:
            nodoActual.padre.der = nodoCambio

        nodoCambio.izq = nodoActual
        nodoActual.padre = nodoCambio

    def __rotarDer(self, x):
        y = x.izq
        x.izq = y.der
        if y.der != self.nil:
            y.der.padre = x
        y.padre = x.padre
        if x.padre is None:
            self.raiz = y
        elif x == x.padre.der:
            x.padre.der = y
        else:
            x.padre.izq = y
        y.der = x
        x.padre = y
```

```python
def count_nodes(self):
        return self.__contar(self.raiz)

def __contar(self, nodo):
        if nodo == self.nil:
            return 0
        return 1 + self.__contar(nodo.izq) + self.__contar(nodo.der)
        
def range_query(self, low, high):
        resultado = []
        self.__range_query(self.raiz, low, high, resultado)
        return resultado

def __range_query(self, nodo, low, high, resultado):
        if nodo == self.nil:
            return
        if low < nodo.clave:
            self.__range_query(nodo.izq, low, high, resultado)
        if low <= nodo.clave <= high:
            resultado.append((nodo.clave, nodo.descripcion))  
        if nodo.clave < high:
            self.__range_query(nodo.der, low, high, resultado)
```

```python
def successor(self, clave):
    nodo_actual = self.__buscar_nodo(clave)
    if nodo_actual == self.nil:
        return None
    
    if nodo_actual.der != self.nil:
        sucesor = self.__minimo_subarbol(nodo_actual.der)
        return (sucesor.clave, sucesor.valor)
    
    sucesor = nodo_actual.padre
    while sucesor != self.nil and nodo_actual == sucesor.der:
        nodo_actual = sucesor
        sucesor = sucesor.padre
    
    if sucesor != self.nil:
        return (sucesor.clave, sucesor.valor)
    else:
        return None
    
def predecessor(self, clave):
    nodo_actual = self.__buscar_nodo(clave)
    if nodo_actual == self.nil:
        return None
    
    if nodo_actual.izq != self.nil:
        predecesor = self.__maximo_subarbol(nodo_actual.izq)
        return (predecesor.clave, predecesor.valor)
    
    predecesor = nodo_actual.padre
    while predecesor != self.nil and nodo_actual == predecesor.izq:
        nodo_actual = predecesor
        predecesor = predecesor.padre
    
    if predecesor != self.nil:
        return (predecesor.clave, predecesor.valor)
    else:
        return None
    
def inorder(self):
    lista = []
    self.__inordenRecursivo(self.raiz, lista)
    return lista

def __inordenRecursivo(self, nodo, resultado):
    if nodo != self.nil:
        self.__inordenRecursivo(nodo.izq, resultado)
        resultado.append((nodo.clave, nodo.valor))
        self.__inordenRecursivo(nodo.der, resultado)
```
